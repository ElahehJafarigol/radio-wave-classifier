```python
import math
import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt
```

<h1>Generate and Plot Two Frequencies</h1>

```python
# Frequency, in cycles per second, or Hertz
f1 = 2  
f2 = 3
# sampling rate, or number of measurements per second
sample_rate = 100
seconds = 10
intervals = seconds * sample_rate

# create signal time series data 
t = np.linspace(0, seconds, intervals)
sin1 = np.sin(f1 * 2 * np.pi * t)
sin2 = np.sin(f2 * 2 * np.pi * t)

# plot signal
plt.plot(t, sin1)
plt.plot(t, sin2)
plt.xlim(0,2)
plt.title('sin(2t) and sin(3t)')
plt.xticks([], [])
plt.yticks([], [])
plt.show()
```

<h1>Combine Sine Waves and Plot Resulting Signal </h1>

```python
# generate signal
signal = []
for i in range(len(sin1)):
    signal.append(sin1[i] + sin2[i])
    
# plot signal
plt.plot(t, signal)
plt.xlim(0,2)
plt.title('Combined Signal')
plt.xticks([], [])
plt.yticks([], [])
plt.show()
```

<h1>Plot Fast Fourier Transform</h1>

```python
signal_fft = fftpack.fft(signal)
signal_fft = np.abs(signal_fft)
freqs = fftpack.fftfreq(len(signal)) * sample_rate

# plot signal
multiplier = 3
plt.figure(figsize=(multiplier*4,multiplier*1.5))
plt.plot(freqs, signal_fft)
plt.xlim(0,5)
plt.grid(True)
plt.rcParams.update({'font.size': 20})
plt.style.use(['dark_background'])
#plt.xticks([], [])
plt.yticks([], [])
plt.show()
```
